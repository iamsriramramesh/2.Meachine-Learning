```python
#TITLE 01:MODEL CREATION/LEARNING
#importing pandas liberary
import pandas as pd
```

```python
#1.DATA COLLECTION
#importing database for model creation with given data from client
database=pd.read_csv("Salary_Data.csv")
```

```python
#database for model creation is imported
database
```

```python
#2.INPUT/OUTPUT SPLIT
#spliting input from database by assigning variable 
independent= database[["YearsExperience"]]
independent
```

```python
#spliting output from database by assigning variable 
dependent= database[["Salary"]]
dependent
```

```python
#3.SPLIT TRAIN AND TEST
#by importing sklearn liberary, train test split function is imported, and assigning parameter with variables
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#4.MODERL CREATION
#creating model through importing linear regression from sklearn liberary, and assigning variable to linear regression as regressor, and giving input to the model(regressor)
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#calculating weight (w) from the model
weight= regressor.coef_
weight
```

```python
#calculating bios (b) from the model
bios= regressor.intercept_
bios
```

```python
#5.EVALUATION METRICS
#to evaluate the model, input test data is evaluated with model and assigned as variable(y_pred) to find r_score 
y_pred= regressor.predict(x_test)
```

```python
#y_pred and output test data is evaluated with r_score formula
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
#if r_score value is near to 1 is good, near to 0 is bad model
r_score
```

```python
#6.saving the model
#importing pickle, and naming the file
import pickle
filename="finalised_model_linear.sav"
```

```python
#dumping the model in the saved file
pickle.dump(regressor,open(filename,'wb'))
```

```python
#TITLE 02:DEPLOYMENT
#loading the model
loaded_model=pickle.load(open("finalised_model_linear.sav",'rb'))
result=loaded_model.predict([[0]])
```
